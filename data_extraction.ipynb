```python
from scipy.io import netcdf
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import netCDF4
import os
import tensorflow as tf
import glob
import pandas as pd
```

```python
our_columns = ['icethic_cea', 'iceconc','uice_ipa', 'vice_ipa','snowthic_cea']

coord_x = list(np.linspace(1, 451, 5, dtype=int))
nc = netCDF4.Dataset('ARCTIC_1h_ice_grid_TUV_20150101-20150101.nc')
nav_lon = np.array(nc.variables['nav_lon'][:, 200])[coord_x]
nav_lat= np.array(nc.variables['nav_lat'][:, 200])[coord_x]
coord = np.array([(lon, lat) for lon,lat in zip(nav_lon,nav_lat)])
print(coord.shape)
path = '/home/user/Downloads/arctic_data'

for columns in our_columns:
    print(columns)
    icethic_cea = []

    
    for foldername in os.listdir(path):
        for filename in glob.glob(os.path.join(path, foldername, '*.nc')):
            temp_array = np.ndarray((5,24))
            nc = netCDF4.Dataset(str(filename))
            target_feature = nc.variables[columns][:, :, 200].data
            counter = 0
            for x in coord_x:
                temp_array[counter] = target_feature[:, x]
                counter += 1
            try:
                icethic_cea = np.hstack((icethic_cea, temp_array))
            except:
                icethic_cea = temp_array.copy()
        print(icethic_cea.shape)
    
    icethic_cea = np.hstack((coord, icethic_cea))
    np.savetxt("/home/user/Downloads/arctic_results/"+columns+".csv", icethic_cea, delimiter=",")
    print(icethic_cea.shape)
```

```python
our_columns = ['icethic_cea', 'iceconc','uice_ipa', 'vice_ipa','snowthic_cea']

coord_x = list(np.linspace(1, 451, 5, dtype=int))
nc = netCDF4.Dataset('ARCTIC_1h_ice_grid_TUV_20150101-20150101.nc')
nav_lon = np.array(nc.variables['nav_lon'][:, 200])[coord_x]
nav_lat= np.array(nc.variables['nav_lat'][:, 200])[coord_x]
coord = np.array([(lon, lat) for lon,lat in zip(nav_lon,nav_lat)])
print(coord.shape)
path = '/home/user/Downloads/arctic_data'
icethic_cea = []
for foldername in os.listdir(path):
    for filename in glob.glob(os.path.join(path, foldername, '*.nc')):
        nc = netCDF4.Dataset(str(filename))
        temp_array = np.ndarray((24,5))
        for cn, column in enumerate(our_columns):
            target_coord = nc.variables[column][:, 200, 200].data
            temp_array[:, cn] = target_coord
        try:
            icethic_cea = np.vstack((icethic_cea, temp_array))
        except:
            icethic_cea = temp_array.copy()
        print(icethic_cea.shape)

np.savetxt("/home/user/Downloads/arctic_results/ts_features.csv", icethic_cea, delimiter=",")
```
